```python
#Function to create an array of marker size for each point, depending on its frequency percentage
def marker_size(variable): 
    """
    variable should be the column of a dataframe for which I wish to create an array of marker sizes
    
    """
    #Create a frequency table 
    base=pd.crosstab(index=variable, columns='count',).sort_values(by='count')

    y=len(variable)

    #Determine a function for the marker size based on the ratio of occurence of the value divided by the 
    #set size: The percentage of occurence of such value in the dataset.
    
    set_size= lambda x:((x/y)*100)*4
    
    #Create a column called size, for the marker size that corresponds to each value
    base['size'] = base.apply(lambda x: set_size(base['count']))
    
    size_array=[]
    
    #create an array with marker sizes for each element in the dataset.
    for each in variable:
        size_array.append(int(base.loc[[each],['size']]['size']))
    return size_array
```

```python
def draw_histograms(bening, malignant, variables, nrows, ncols):
    fig,axes=plt.subplots(nrows,ncols,sharex=True, figsize=(20,20))
    c=[0,0,'#0000FF', 0, '#FF0000']
    color=[c[x] for x in data['class']]

    for i, var in enumerate(variables):
        axes[4,i].set_xlabel(var)
        axes[i,0].set_ylabel(var)
        y = malignant[var]

        for j, var2 in enumerate(variables):
            x = benign[var2]
            
            if j==i:
                axes[i,j].hist(x, bins=20, alpha=0.55, label='Benign',color=c[2])
                axes[i,j].hist(y, bins=20, alpha=0.55, label='Malign',color=c[4])
            else:
                # size = 120
                size=np.array([[marker_size(data[variables[i]])],[marker_size(data[variables[j]])]])
                axes[i,j].scatter(data[variables[j]],data[variables[i]], alpha=0.3, color=color, s=size)
        
    fig.text(0.5,0.04,'Variables', ha='center', va='center', fontsize=30)
    fig.text(0.05,0.5,'Frequency', ha='center', va='center', \
                 rotation='vertical',fontsize=30)
    
    hndl = [Line2D([0], [0], marker='.', color=c[2], label='Scatter',markerfacecolor=c[2], markersize=15),\
          Line2D([0], [0], marker='.', color=c[4], label='Scatter',markerfacecolor=c[4], markersize=15)]
    
    fig.legend(handles=hndl,labels=('Bening','Malignant'),loc='upper right',prop={'size': 16})
    plt.show()
```

```python
# read the whole data
import pandas as pd
from sklearn.decomposition import PCA
whole = pd.read_excel('breast-cancer-wisconsin.xlsx', 0, \
                     index_row=0, na_values=[''])
whole = whole.interpolate(method='linear')

pca = PCA(n_components=9)
# disconsider the first column (code) and the last (class)
pca.fit(whole.iloc[:,1:10])
#print(.head())
#print(pca.components_)

ratios = np.cumsum(pca.explained_variance_ratio_)
print(ratios)
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.plot(ratios)
plt.axes().set_ylim(0, 1.1)
plt.axhspan(0.895,0.9, facecolor='g', alpha=0.4)
plt.show()
```
